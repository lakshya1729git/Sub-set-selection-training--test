```
# This is formatted as code#include <stdio.h>
#include <stdlib.h>
#include <time.h>

// Function to allocate memory for a matrix dynamically
int** allocateMatrix(int rows, int cols) {
    int** matrix = (int**)malloc(rows * sizeof(int*));
    for (int i = 0; i < rows; i++) {
        matrix[i] = (int*)malloc(cols * sizeof(int));
    }
    return matrix;
}

// Function to free memory for a matrix
void freeMatrix(int** matrix, int rows) {
    for (int i = 0; i < rows; i++) {
        free(matrix[i]);
    }
    free(matrix);
}

int main() {
    int r1, c1, r2, c2;

    // Input dimensions of matrices
    printf("Enter rows and columns of first matrix: ");
    scanf("%d %d", &r1, &c1);
    printf("Enter rows and columns of second matrix: ");
    scanf("%d %d", &r2, &c2);

    // Check if multiplication is possible
    if (c1 != r2) {
        printf("Matrix multiplication not possible with these dimensions.\n");
        return 1;
    }

    // Allocate memory for matrices
    int** a = allocateMatrix(r1, c1);
    int** b = allocateMatrix(r2, c2);
    int** res = allocateMatrix(r1, c2);

    // Seed random number generator
    srand(time(0));

    // Fill first matrix with random values between 1 and 10
    printf("Filling first matrix with random values...\n");
    for (int i = 0; i < r1; i++) {
        for (int j = 0; j < c1; j++) {
            a[i][j] = (rand() % 10) + 1; // Random values between 1 and 10
        }
    }

    // Fill second matrix with random values between 1 and 10
    printf("Filling second matrix with random values...\n");
    for (int i = 0; i < r2; i++) {
        for (int j = 0; j < c2; j++) {
            b[i][j] = (rand() % 10) + 1; // Random values between 1 and 10
        }
    }

    // Start measuring time
    clock_t start = clock();

    // Matrix multiplication logic
    for (int i = 0; i < r1; i++) {
        for (int j = 0; j < c2; j++) {
            res[i][j] = 0;
            for (int k = 0; k < c1; k++) {
                res[i][j] += a[i][k] * b[k][j];
            }
        }
    }

    // Stop measuring time
    clock_t end = clock();
    double time_taken = ((double)(end - start)) / CLOCKS_PER_SEC;

    // Display time taken
    printf("Time taken for matrix multiplication of size %dx%d and %dx%d: %f seconds\n", r1, c1, r2, c2, time_taken);

    // Write results to CSV file
    FILE* file = fopen("matrix_results.csv", "a");
    if (file == NULL) {
        printf("Error opening file.\n");
        return 1;
    }
    fprintf(file, "Matrix A: %dx%d, Matrix B: %dx%d, Time Taken: %f seconds, Language: C\n", r1, c1, r2, c2, time_taken);
    fclose(file);

    // Free allocated memory
    freeMatrix(a, r1);
    freeMatrix(b, r2);
    freeMatrix(res, r1);

    return 0;
}

    chrono::duration<double> time_taken = end - start;

    // Free allocated memory

```
# This is formatted as codeimport random
import time
import csv
import numpy as np  # Importing numpy for matrix operations

# Function to perform matrix multiplication using NumPy
def matrix_multiply(a, b):
    return np.dot(a, b)

def get_matrix_dimensions():
    while True:
        try:
            r, c = map(int, input("Enter number of rows and columns (separated by space): ").split())
            if r <= 0 or c <= 0:
                print("Rows and columns must be positive integers. Try again.")
                continue
            return r, c
        except ValueError:
            print("Invalid input! Please enter two integers separated by space.")

def main():
    # Get matrix dimensions
    print("For Matrix A (rows and columns):")
    r1, c1 = get_matrix_dimensions()
    print("For Matrix B (rows and columns):")
    r2, c2 = get_matrix_dimensions()

    # Check if multiplication is possible
    if c1 != r2:
        print("Matrix multiplication not possible with these dimensions.")
        return

    # Generate random matrices with values between 1 and 10 using NumPy
    print("Filling matrices with random values...")
    a = np.random.randint(1, 11, size=(r1, c1))  # Random matrix A
    b = np.random.randint(1, 11, size=(r2, c2))  # Random matrix B

    # Start measuring time for matrix multiplication
    start_time = time.time()

    # Perform matrix multiplication using NumPy's dot function
    result = matrix_multiply(a, b)

    # Stop measuring time
    end_time = time.time()

    # Calculate time taken for multiplication
    time_taken = end_time - start_time

    # Display time taken
    print(f"Time taken for matrix multiplication of size {r1}x{c1} and {r2}x{c2}: {time_taken} seconds")

    # Store data in CSV file
    with open('matrix_results.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([f"Matrix A: {r1}x{c1}", f"Matrix B: {r2}x{c2}", f"Time Taken: {time_taken} seconds", "Language: Python with NumPy"])

if __name__ == "__main__":
    main()

Importing csv file and creating graphs using pandas and google colab features


````python
import pandas as pd
import re
from google.colab import files

# Step 1: Upload the CSV file
uploaded = files.upload()

# Read the uploaded CSV file (assuming the uploaded file is named 'matrix_results.csv')
file_name = list(uploaded.keys())[0]  # Get the file name of the uploaded file
df = pd.read_csv(file_name)

# Step 2: Extract matrix dimensions and time taken for analysis
def extract_matrix_dimensions(matrix_str):
    try:
        # Match the pattern of 'number x number' like '1000x1000'
        match = re.search(r'(\d+)x(\d+)', matrix_str)
        if match:
            return int(match.group(1)), int(match.group(2))
        else:
            return None, None  # Return None if the format is incorrect
    except Exception as e:
        return None, None  # Return None if there's an exception

# Apply the function to extract rows and columns for both matrices
df["Matrix A Rows"], df["Matrix A Columns"] = zip(*df["Matrix A"].apply(extract_matrix_dimensions))
df["Matrix B Rows"], df["Matrix B Columns"] = zip(*df["Matrix B"].apply(extract_matrix_dimensions))

# Step 3: Display data in tabular form
df
````
